```python
from pyspark import SparkConf, SparkContext
conf = SparkConf().setMaster('local').setAppName('coustomerTotalSpent')
sc = SparkContext(conf = conf)
```

```python
def parseLine(line):
    fields = line.split(',')
    customerId = int(fields[0])
    amount = float(fields[2])
    return(customerId, amount)
```

```python
lines = sc.textFile('customer-orders.csv')
parsedLine = lines.map(parseLine) 
amtSpentByCustomer = parsedLine.reduceByKey(lambda x,y : x+y)
results = amtSpentByCustomer.sortBy(lambda x : x[1], False).collect()
for result in results:
    print('{0} \t {1:.2f}'.format(result[0], result[1]))
```

```python
lines = sc.textFile("customer-orders.csv")
orders = lines.map(parseLine)
customers = orders.reduceByKey(lambda x,y : x + y)
amounts = customers.map(lambda x : (x[1],x[0]))
result = amounts.sortByKey(False).collect()

for x in result:
    print (x)
```
